# Complexity Metric Problematization - Examples

## P1 Monotone growth

The complexity metric grows with increasing window size, although there is no variance, rare species or loops.

```python
from pm4py.objects.log.obj import EventLog, Trace, Event

# Create a trace variant A-B-C-D-E
variant = ["A", "B", "C", "D", "E"]

def make_trace(variant):
    t = Trace()
    for act in variant:
        t.append(Event({"concept:name": act}))
    return t

# Build EventLog with 10,000 identical traces
event_log = EventLog([make_trace(variant) for _ in range(10000)])
```

```python
from complexity_sample_size_correlation_analysis import compute_metrics_for_samples, sample_random_traces_with_replacement

sizes = range(10, 501, 50)
samples_per_size = 10
random_state = 1
samples = sample_random_traces_with_replacement(event_log, sizes, samples_per_size, random_state)
adapters = ["vidgof_sample", "population_inext", "population_simple"]
df_metrics = compute_metrics_for_samples(samples, adapters)
df_metrics
```
